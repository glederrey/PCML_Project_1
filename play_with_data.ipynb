# Playing with the data

In this notebook, we just want to play a bit the data to understand a bit what they are about. First, we load the train data with pandas.

```python
"""
All the values with -999 are considered as NaN values. Therefore, we 
replace them directly with NaN values for the later computation.
"""
train = pd.read_csv('data/train.csv', na_values = "-999") 
train
```

We can now drop the column *Id*. The index column is sufficient. 

```python
train = train.drop('Id', axis=1)
train.head(10)
```

Now, we just check the NaN values for each column. It's just to make sure that the *Prediction* column has no NaN value.

```python
# False implies there's no NaN value.
train.isnull().any() 
```

We see in this case that some of the columns have NaN values. So, these parameters will surely have a lesser weight for the prediction.

Now, we will separate the two predictions. Then, we'll be able to do some stats on these two predictions.


```python
# Print the predictions to make sure we have only two kind of predictions
pred = np.unique(train.Prediction)
print("The predictions are: %s"%(", ".join(str(p) for p in pred)))

# Now, we use the Prediction as the index on the right.
prediction = train.Prediction
train.index = prediction
# Drop the prediction column since it's the new index
train = train.drop('Prediction', axis=1)
train.head()
```

```python
# Count the number of object s
sum = 0
for i in range(len(pred)):
    sum = sum + train.loc[pred[i]].count(axis=0).DER_deltar_tau_lep
    print("Number of prediction with %s: %i"%(pred[i],
                train.loc[pred[i]].count(axis=0).DER_deltar_tau_lep))
    
print("Total: %i"%sum)
```

Now that we know the predictions, we can compare the different columns between the two predictions.

```python
for col in train.columns:
    dic = {}
    for p in pred:
        dic[p] = train.loc[p, col].describe()

    val = pd.DataFrame(dic)
    print(col)
    print(val)
    print()
    
    
```
